```python
import numpy as np
```

```python

losess=list()
for c1 in np.arange(0.3,1.6,0.05):
    for c2 in np.arange(0.3,1.6,0.05):
        for i in np.arange(0.3,1.6,0.05):
            for r1 in np.arange(0.3,1.6,0.05):
                y = -c1*c2**2*(-4*i+c1*r1**2)
                if y<0:
                    print("L1, L2 - complex " +str(c1)+" "+str(r1)+" "+str(i)+" "+str(c2))
                    losess.append([c1,r1,i])
                    break
    
```

```python

losess2=list()
for c1 in np.arange(0.3,1.6,0.05):
    for c2 in np.arange(0.3,1.6,0.05):
        for i in np.arange(0.3,1.6,0.05):
            for r2 in np.arange(0.3,1.6,0.05):
                y = -c2*c1**2*(-4*i+c2*r2**2)
                if y<0:
                    print("L1, L2 - complex " +str(c1)+" "+str(r2)+" "+str(i)+" "+str(c2))
                    losess2.append([c2,r2,i])
                    break
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d

x = np.array(losess)[:,0]
y = np.array(losess)[:,1]
X, Y = np.meshgrid(x, y)
Z= np.array(losess)[:,2].reshape(1,-1)
print(Z)
fig = plt.figure()
fig.set_figheight(10)
fig.set_figwidth(13)
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_xlabel('Hydraulic resistance')
ax.set_ylabel('Ellastic coefficient')
ax.set_zlabel('Inertion parametr')
ax.set_title('Plot')
plt.show()
```
